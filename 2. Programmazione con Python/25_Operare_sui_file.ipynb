1. Filtra i proverbi

Leggi il contenuto del file proverbi.txt e scrivi su di un nuovo file chiamato proverbi_filtrati.txt solo i proverbi che iniziano per vocale o che sono più brevi di 25 caratteri

```python
from google.colab import files
uploaded = files.upload()
```

```python
with open("proverbi.txt") as proverbs_file:
  proverbs_filtrati = open("proverbs.filtrati.txt", "w")

  for proverb in proverbs_file.readlines():  #con readlines itera su ogni riga
    if len(proverb)< 25 or proverb[0].lower() in "aeiou":  #oppure poteva essere ["a", "e", "i", "o", "u"]
      proverbs_filtrati.write(proverb)
```

2. Analisi di magazzino

Apri il file shirts.csv e calcola le seguenti statistiche su di esso:

1. Numero totale di prodotti esistenti
2. Valore totale del magazzino (la somma del prezzo di tutti i prodotti)
3. Prezzo medio dei prodotti
4. Numero di prodotti per ogni colore
5. Numero di prodotti per ogni taglia

Sava le statistiche all'interno di un file json chiamato shirts_stats.josn

```python
from google.colab import files
uploaded = files.upload()
```

```python
import csv

with open("shirts.csv") as csv_shirts:
  shirts_reader = csv.reader(csv_shirts)
  next(shirts_reader)

#inizializzo tutti i contatori a zero
  numero_prodotti = 0
  valore_prodotti = 0
  colore_prodotti = {}
  taglia_prodotti = {}


#con questo gli sto dicendo di contare ogni riga
  for i, row in enumerate(shirts_reader):
    numero_prodotti+=1

#qui gli sto dicendo che il prezzo si trova alla terza colonna e che ad ogni riga, se trova il prezzo deve aggiungerlo al contatore
    prezzo = float(row[3])
    valore_prodotti= valore_prodotti + prezzo

#qui gli sto dicendo che il colore si trova alla secondo colonna e che ad ogni riga, se il colore esiste, aggiunge una unità di conto al valore della chiave,
    colore = row[2]  #se non esiste deve creare la chiave
    if colore in colore_prodotti:
      colore_prodotti[colore] +=1
    else:
      colore_prodotti[colore] = 1


    taglia = row[1]
    if taglia in taglia_prodotti:
      taglia_prodotti[taglia] +=1
    else:
      taglia_prodotti[taglia] = 1


mean_price = valore_prodotti/numero_prodotti

statistiche = {
    "numero_prodotti": numero_prodotti,
    "valore_totale": valore_prodotti,
    "prezzo_medio": mean_price,
    "colore_prodotti": colore_prodotti,
    "taglie_prodotti": taglia_prodotti
}

import json

with open("shirts_stats.json", "w") as json_file:
    json.dump(statistiche, json_file, indent=4)


print(statistiche)
```
