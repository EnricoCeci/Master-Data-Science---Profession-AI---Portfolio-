1. Vettori Reloaded

Durante la nostra esercitazione con le classi abbiamo realizzato una classe per rappresentare vettori, in più metodi abbiamo verificato che la lunghezza di due vettori fosse uguale. Rimpiazziamo l'if utilizzato in questi casi con un assert.

```python
class Vector():

    def __init__(self, values):
        self.values = values


    def __getitem__(self, i):
      return self.values[i]


    def __repr__(self) :
       return str(self.values)


    def __add__(self, vector):

      assert(len(self.values)==len(vector.values)), "i vettori hanno lunghezza diversa"

      list_sum = []

      for val1, val2 in zip(self.values, vector.values):
        list_sum.append(val1+val2)

      return Vector(list_sum)


      def __sub__(self, vector):

        assert(len(self.values)==len(vector.values)), "i vettori hanno lunghezza diversa"

        list_sum = []

        for val1, val2 in zip(self.values, vector.values):
          list_sum.append(val1-val2)

        return Vector(list_sum)


      def __eq__(self, vector):

        assert(len(self.values)==len(vector.values)), "i vettori hanno lunghezza diversa"

        for val1, val2 in zip(self.values, vector.values):
          if(val1 != val2):
            return False
        return True


      def __mul__(self, vector):
        return self.dot(vector)


      def sum(self):

        v_sum = 0

        for v in self.values:
          v_sum+=v

        return v_sum


      def dot(self, vector):

        assert(len(self.values)==len(vector.values)), "i vettori hanno lunghezza diversa"

        prod = 0

        for val1, val2 in zip(self.values, vector.values):
          prod+=val1*val2

        return prod


      def norm(self):
        return self.dot(self)**(1/2)

v1 = Vector([1, 2, 3])
v2 = Vector([2, 4, 6, 8])

v1+v2


```

2. Calcolatrice

Definisci una funzione per eseguire dei semplici calcoli, la funzione prende in ingresso due valori numerici ed una stringa contenente un'operazione aritmetica (+, -, *, /) ed esegue tale operazione tra i due numeri. Acquisisci in input i parametri della funzione

1. utilizza il try/except per gestire l'eccezione che si causa nel caso in cui viene inserito un valore non numerico.
2. utilizza l'assert per assicurarti che l'operatore aritmetico sia valido, cioè sia un carattere tra +,-,*,/.
3. utilizza il try/except per gestire il caso di divisioni per 0, in tal caso stampa semplicemente un messaggio 'Non puoi dividere per 0'.

```python
operatori = ("+", "-", "*", "/")

def calcolatrice(num1, num2, operatore):
    assert operatore in operatori, "Questo non è uno degli operatori matematici"
    try:
        if operatore == "+":
            risultato = num1 + num2
        elif operatore == "-":
            risultato = num1 - num2
        elif operatore == "*":
            risultato = num1 * num2
        elif operatore == "/":
            risultato = num1 / num2
        return risultato
    except ZeroDivisionError:
        print("Non puoi dividere per 0")
        return None

# ---- Parte input con gestione valori non numerici ----
try:
    num1 = float(input("Inserisci il primo numero: "))
    num2 = float(input("Inserisci il secondo numero: "))
    operatore = input("Inserisci l'operatore (+, -, *, /): ")

    risultato = calcolatrice(num1, num2, operatore)
    if risultato is not None:
        print("Risultato:", risultato)

except ValueError:
    print("Hai inserito un valore non numerico")
```

In questa esercitazione imparerai a gestire le eccezioni più comuni in Python utilizzando il blocco try-except.

Dovrai scrivere un programma che legge due numeri in input dall'utente, li converte in interi e calcola la loro divisione. Il programma deve gestire in modo corretto le seguenti situazioni: input non numerico (ValueError) e divisione per zero (ZeroDivisionError).

Se uno di questi errori si verifica, il programma deve stampare un messaggio di errore specifico. Inoltre, indipendentemente dal risultato o dagli errori, deve essere stampato un messaggio finale che indica la conclusione dell'operazione (blocco finally).

Usa esclusivamente le strutture viste nel modulo sulle eccezioni (try, except, finally) e le funzioni di input/output già conosciute.

Non è richiesta la gestione di altre eccezioni.

```python
num1 = float(input("Inserisci il primo numero: "))
num2 = float(input("Inserisci il secondo numero: "))

def calcolatrice(num1, num2):
  try:
    risultato = num1 / num2
    print("Il risultato della divisione è:", risultato)


  except ValueError:
    print ("Il valore inserito non è un numero")
    return None

  except ZeroDivisionError:
      print("Non puoi dividere per 0")
      return None



  finally:
    print ("Operazione eseguita")
```

4. Deep Training

In questa esercitazione lavorerai con il modulo delle eccezioni in Python. Ti chiederemo di scrivere un programma che:

1. Riceve in input una linea di testo contenente esattamente due numeri interi separati da spazi.

2. Controlla che l'input contenga due e solo due numeri interi. Se la verifica fallisce, il programma deve sollevare (raise) una eccezione ValueError con un messaggio personalizzato "Devi inserire esattamente due numeri interi".

3. Se l'input è valido, esegue la divisione del primo numero per il secondo e stampa il risultato.

4. Gestisce eventuali eccezioni generate, nel dettaglio:
   - ValueError per input non corretto o conversione fallita;
   - ZeroDivisionError per divisione per zero;
   - qualsiasi altra eccezione deve essere intercettata da un except generico.

Nel programma dovrai utilizzare il costrutto try-except per gestire le eccezioni. Ricordati di usare l'alias "as" per poter mostrare il messaggio di errore fornito dall'eccezione stessa.

Inoltre, usa un blocco finally per stampare un messaggio di chiusura, ad esempio "Esecuzione del programma terminata." sempre al termine del programma, indipendentemente dal risultato.

Questo esercizio ti permetterà di applicare le conoscenze acquisite su try-except-finally, sollevare eccezioni con raise e gestire input non valido in modo controllato, mantenendo il programma robusto e con messaggi informativi.

```python
try:  #prova ad effettuare la divisione che prenda in input due numeri separati da uno spazio che verrano divisi con il comando sotto
    numeri = list(map(int, input("Inserisci due numeri interi separati da spazio: ").split()))
    if len(numeri) != 2:   #MA prima di eseguire la divisione verifica che in numeri inseriti siano solo 2
        print("Devi inserire esattamente due numeri interi")
        raise ValueError

    a, b = numeri   #verificato che la lunghezza sia corretta vai con la divisione e...
    c = a / b
    print(f"Il risultato della divisione è: {c}")  #printa il risultato eccetto nei casi in cui vediamo sotto

#non printare il risultato  nel caso in cui il divisore fosse 0
except ZeroDivisionError as e:   #"e" serve per stampare il messaggio di default dell'eccezione
    print(e)
    c = None
#non printare il risultato neanche se il valore inserito non è un numero
except ValueError as e:
    print(e)
    c = None
#Inoltre non printare il risultato per qualunque altro tipo di errore
except Exception as e:  #per riferirci ad un'eccezione generica usiamo il generico Exception
    print(e)
    c = None

finally:
    print("Operazione eseguita")
```
